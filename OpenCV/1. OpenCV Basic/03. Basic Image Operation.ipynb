### 1. Read, Write and Display an Image

OpenCV에서 imread는 이미지를 읽는 데 사용되고 imwrite는 메모리에 이미지를 쓰고 imshow는namedWindow와 함께 사용되며 waitKey는 메모리에 이미지를 표시하는 데 사용됩니다.

Jupyter Notebook 내에서 이미지를 표현하려면 matplotlib.pyplot을 이용하여 표현할 수 있습니다.
cv2.imshow()를 이용하면 새 창 생성되고 그 창에서 이미지가 나타나는 반면 matplotlib을 이용하면 Jupyter notebook에 좀더 자연스럽게 표현할 수 있습니다.

```python
# Import Opencv & matplotlib
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the image
image = cv2.imread('images/ioniq.jpg', 1)

# Check for invalid input
if image is None:
    print('Could not open or find the image')

# Save the image data into the Numpy
print(type(image))
```

```python
# Convert Color image to grayscale
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Save the result
cv2.imwrite('image/gray_ioniq.jpg', grayImage)
```

```python
# Display the image
cv2.imshow("original", image)
cv2.imshow("grayscale", grayImage)
```

```python
# Wait for a keystroke in the winsow

# waitKey parameter take the N miliseconds to show the image in the windhow.
# if parameter is 0, image will be shown until any key is pressed.
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Display the image in Jupyter notebook

# Convert the BGR image(OpenCV) to RGB (matplotlib)
RGB_origin_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_origin_image)
plt.show()
```

### 2. Crop and Resize

OpenCV에서는 이미지의 크기를 조정할 수 있습니다. 이미지를 축소하고 확대 할 수 있습니다. 즉, 이미지는 특정 작업을 수행하기 위해 축소 된 다음 작업이 완료되면 다시 원본 크기로 확대할 수 있습니다. 이미지 크기 축소는 계산 속도를 향상시키는 데 도움이 됩니다. 이미지를 잘라낼 수도 있습니다. 이는 전체 이미지를 처리 할 필요가 없기 때문에 계산 속도를 높이는 데 도움이됩니다.

아래 코드는 OpenCV의 크기 조정 기능을 사용하여 이미지를 축소하고 확대하는 방법을 보여줍니다. 그런 다음 범위 기능을 사용하여 이미지를 잘라내어 이미지의 일부를 추출합니다.

다시 한번 정리하면 <span class="mark">resize를 하는 목적</span>은 다음과 같습니다.
1. 이미지 처리 연산 속도 증가
2. 이미지 비율 변경

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

The resize function has the following prototype:

destinationImage =  cv2.resize(sourceImage, (,), 
                               scale factor in x direction, 
                               scale factor in y direction, 
                               interpolation method
                               )
                               
                               
출력의 높이와 너비를 결정하기 위해 출력의 크기를 지정할 수도 있고, 스케일링 팩터를 추가 할 수도 있고 크기가 자동으로 계산됩니다.

```python
image = cv2.imread("image/ioniq.jpg")
scaleX = 0.6
scaleY = 0.6

# Scaling down the image 0.6 times
scaleDown = cv2.resize(image, None, fx = scaleX, fy = scaleY, interpolation = cv2.INTER_LINEAR)

# Scaling up the image 1.8 times
scaleUp = cv2.resize(image, None, fx = scaleX * 3, fy = scaleY * 3, interpolation = cv2.INTER_LINEAR)
```

```python
plt.imshow(scaleUp)
plt.show()
```

Crop을 위해 numpy 배열의 slicing 기능을 직접 사용할 수 있으며, 먼저 slicing할 영역의 행과 열을 지정하여 슬라이스합니다.

```python
# Cropped Image
crop = image[300:350, 200:250]

# Display all images
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
scaleDown = cv2.cvtColor(scaleDown, cv2.COLOR_BGR2RGB)
scaleUp = cv2.cvtColor(scaleUp, cv2.COLOR_BGR2RGB)
crop = cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)

plt.subplot(221)
plt.title("original"); plt.imshow(image)
plt.subplot(222)
plt.title("Scale down"); plt.imshow(scaleDown)
plt.subplot(223)
plt.title("Scale up"); plt.imshow(scaleUp)
plt.subplot(224)
plt.title("Crop"); plt.imshow(crop)

plt.tight_layout()      
```

### 3. Datatype conversion

OpenCV를 이용하여 imread를 할 때 특별히 date type을 지정하지 않으면 unsigned int(8bit)로 저장되게 됩니다. Computer Vision에서는, 이미지를 32 비트 데이터를 갖는 각 픽셀과 함께 부동 소수점 데이터 유형으로 변환합니다. 이렇게하면 이미지 작업 중 **오버 플로우를 방지** 할 수 있습니다. 

float point로 <span class="mark">Data conversion하는 이유</span>를 정리하면 다음과 같습니다.
1. 연산 시 overflow 방지
2. 연산 시 numerical accuracy 향상

<span class="mark">range를 0~1로 변환</span>하는 이유
1. normaliza 하기 위함
2. imshow 함수 이용 시 uint8에 대응하는 범위(0 ~ 255)에 맞추기 위함

```python
import cv2
import numpy as np

# Read image
image = cv2.imread("images/ioniq.jpg")
print(type(image[0][0][0]))
```

```python
scalingFactor = 1/255.0
# Conver unsigned into float
image = np.float32(image)
image = image * scalingFactor
print(type(image[0][0][0]))
```

경계의 두께는 thickness 매개 변수에 의해 제어됩니다. 기본적으로 모양의 윤곽선 만 그려집니다. 지정된 색상으로 채워진 모양을 그리려면 두께가 -1이어야합니다. <br>
lineType 매개 변수는 렌더링 품질을 제어합니다. lineType이 CV_LINE_AA로 설정되면 앤티 앨리어싱 된 (부드러운) 선이 그려집니다

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read a image.
image = cv2.imread('images/initial_D.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Draw a line
imageLine = image.copy()
cv2.line(imageLine, (300, 120), (600, 120), (255, 0, 0), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(imageLine)
plt.show()
```

```python
# Draw a circle
imageCircle = image.copy()
cv2.circle(imageCircle, (330, 430), 80, (255, 0, 0), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(imageCircle)
plt.show()
```

```python
# Draw an ellipse
# : Ellipse Centers and Axis lengths must be integers

imageEllipse = image.copy()
cv2.ellipse(imageEllipse, (320, 420), (50, 110), 0, 0, 360, (255, 0, 0), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imageEllipse)
plt.show()
```

```python
# Draw a rectangle
imageRectangle = image.copy()
cv2.rectangle(imageRectangle, (20, 100), (900, 530), (255, 0, 0), thickness = 5, lineType = cv2.LINE_AA)
cv2.rectangle(imageRectangle, (520, 10), (760, 550), (0, 0, 250), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(imageRectangle)
plt.show()
```

```python
# Put text into image
cv2.putText(imageRectangle, "Car", (20, 90), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 0, 0), thickness = 5, lineType = cv2.LINE_AA)
cv2.putText(imageRectangle, "Person", (300, 50), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(imageRectangle)
plt.show()
```
