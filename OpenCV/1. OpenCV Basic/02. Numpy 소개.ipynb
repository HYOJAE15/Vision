## 3. Numpy in OpenCV


### 3.1 Read an image

OpenCV-Python 환경에서 cv2.imread () 함수를 사용하여 이미지를 읽습니다. 이미지는 작업 디렉토리에 있거나 이미지의 전체 경로가 주어져야합니다. 첫 번째 인수는 파일 이름이고, 두 번째 인수는 이미지를 읽어야하는 방법을 지정하는 플래그입니다.

- cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag.
- cv2.IMREAD_GRAYSCALE : Loads image in grayscale mode
- cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel

각각을 숫자로 옵션을 구성할 수 있습니다. Color = 1, Grayscale = 0, Unchanged = -1이 됩니다.

아래의 Python 코드에서 이미지 파일 "bat.jpg"를 저장했습니다. 그런 다음 그레이 스케일로 이미지를 로드하려고 하므로 두 번째 인수로 0을 전달합니다

```python
# Import numpy ans OpenCV
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# load a color image in grayscale
img = cv2.imread('image/ioniq.jpg', 0)

# load original image
origin = cv2.imread('image/ioniq.jpg', 1)

# Print the image.
# Convert Color type BGR(OpenCV) to RGB(pyplot)
origin = cv2.cvtColor(origin, cv2.COLOR_BGR2RGB)
plt.imshow(origin)
plt.show()
```

### 3.3 Copy part of an image

때때로 우리는 전체 이미지가 아닌 이미지의 일부를 복사해야합니다. 아래 예제는 직사각형 영역을 잘라내는 방법을 보여줍니다.

```python
# Import numpy ans OpenCV
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# Load a color image in grayscale
img = cv2.imread('image/ioniq.jpg', 1)

# Copies from rows 100 to 150 and columns 100 to 250
cropped = img[300:400, 50:200]

# Print the image.
# Convert Color type BGR(OpenCV) to RGB(pyplot)
cropped = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)
plt.imshow(cropped)
plt.show()
```

### 3.4 Find number of rows, columns, and channels

numpy 배열 객체의 shape 메소드는 이미지의 행 수 (높이), 열 수 (너비) 및 채널 수를 찾는 데 사용할 수 있습니다.

```python
# Import numpy and OpenCV
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

# Load a color image in BGR
img = cv2.imread('image/ioniq.jpg', 1)

# Find the number of rows in the image
imgRows = img.shape[0]

# Find the number of columns in the image
imgCols = img.shape[1]

# Find the numer of Channels in the image
Channels = img.shape[2]

print(imgRows, imgCols, Channels)
```
